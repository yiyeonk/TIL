**web에서 사용하는 언어**
* 서버와 클라이언트 간에 데이터를 주고 받을 때, 사용하는 언어

- XML(Extensible Markup Language)
- HTML(HyperText Markup Language)
- 클라이언트(페이지 요청, www.naver.com) -> 웹서버 -> 메인페이지 제공(index.html)

> (1) 구조화된 문서(XML) : 정적 페이지   
    - 웹페이지 내용에 대해 기계가 해석 가능한 문서 -> 검색어의 의미를 기반으로 검색 => 검색 폭 넓음, 검색 결과에 대한 정확도 높음
    - ex) <가수>       <그룹명>BTS</그룹명> ....  </가수>  
(2) 비구조화된 문서(HTML) : 정적 페이지  
    - 웹페이지 내용에 대해 기계가 해석하지 못하는 문서 -> 검색어를 기반으로 검색
    - ex) BTS가 서울 강남구에서 공연을 했습니다.
(3) jsp, asp, php 등 : 동적 페이지(웹 브라우저가 해석 못함)  
- 클라이언트(날씨 클릭) -> 웹서버(날씨 페이지(동적=jsp) 생성) -> 생성된 페이지를 html문서로 만들어 제공 -> 웹브라우저 해석 -> 결과를 화면에 출력

```python
# parsing 원하는 정보를 가져오는 것

from bs4 import BeautifulSoup

html = """
<html>
<body>
<h1>스크래핑</h1>
<p>웹 페이지 분석</p>
<p>원하는 부분 추출</p>
</body>
</html>
"""

# BeautifulSoup(특성,)
soup = BeautifulSoup(html, "html.parser") # ( , 분석기)
# 클래스(대문자 시작)

print(soup.html.body.h1)
print(soup.html.body.p.string)  # string : tag 안 문자열만 출력

p1 = soup.html.body.p
p2 = p1.next_sibling  # 줄바꿈 문자
print(p2.next_sibling.string)
# 위의 식을 간결하게 표현한 식
p2 = p1.next_sibling.next_sibling.string
print(p2)
```

```python
# 1. IT기업 코딩 테스트문제
# 주어진 문자열(공백 없이 쉼표로 구분되어 있음)을 가지고 아래 문제에 대한 프로그램을 작성하세요.

name = "이유덕,이재영,권종표,이재영,박민호,강상희,이재영,김지완,최승혁,이성연,박영서,박민호,전경헌,송정환,김재성,이유덕,전경헌"

# 1-1. 김씨와 이씨는 각각 몇 명 인가요?
kim = len(re.findall("[김]+\w+",name))
lee = len(re.findall("[이]+\w+",name))
print("김씨는 %d명, 이씨는 %d명 입니다. "%(kim, lee))

# 1-2. "이재영"이란 이름이 몇 번 반복되나요?
ljy = len(re.findall('이재영',name))
print("'이재영'은 %d번 반복됩니다."%ljy)

# 1-3. 중복을 제거한 이름을 출력하세요.
name = re.findall("[가-힣]+",name)
print(set(name))

# 1-4. 중복을 제거한 이름을 오름차순으로 정렬하여 출력하세요.
name = re.findall("[가-힣]+",name)
print(sorted(set(name)))
```
